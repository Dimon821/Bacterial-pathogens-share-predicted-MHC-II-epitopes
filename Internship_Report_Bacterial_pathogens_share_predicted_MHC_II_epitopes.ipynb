# 1. Abstract

<div style="text-align: justify; margin: 5px opx; max-width: 1800px;">
Cross-reactivity through CD4+ T cell mediated responses can protect humans from multiple pathogens. In this project, the extent of potential cross-reactivity in four bacterial respiratory pathogens was examined. This was done by slicing peptides from the proteomes, searching them for identical peptides and predicting the peptides's MHC-II binding with NetMHCIIpan. Additionally, the overlap and the predicted MHC-II binding was correlated to detect potential selective pressure to mutate shared epitopes. The found correlations were too weak to indicate a selective pressure. We identified several promising candidate peptides shared across all four studied bacterial species for further investigation in peripheral blood mononuclear cells.
</div>

<p>The adaptive immune system relies on the precise recognition of foreign antigens to coordinate targeted responses against invading pathogens, with CD4+ T helper cells executing a central regulatory role upon interacting with peptides presented by Major Histocompatibility Complex Class II (MHC-II) molecules. The classical MHC-II presentation pathway initiates with the internalization of extracellular pathogens or proteins into endolysosomal compartments, where endogenous proteases cleave these antigens into short peptide fragments. Parallel to antigen processing, the alpha ($\alpha$) and beta ($\beta$) polypeptide chains of the MHC-II heterodimer are co-translationally translocated and assembled within the membrane of the rough endoplasmic reticulum (RER). Within the RER lumen, the nascent MHC-II complex associates with the trimeric invariant chain (Ii, CD74), a molecular chaperone whose core region—the Class II-associated invariant chain peptide (CLIP)—occupies the peptide-binding groove to prevent premature ligand binding. The invariant chain also encodes an endosomal targeting signal that directs the complex from the trans-Golgi network to late endosomal compartments, collectively termed the MHC class II compartment (MIIC). Within the MIIC, sequential proteolytic cleavage degrades the invariant chain, leaving only the CLIP fragment bound to the groove. Subsequent fusion of the antigen-containing endolysosomes with the MIIC enables the exchange of CLIP for a high-affinity antigenic peptide, a process catalyzed by the non-classical, specialized HLA-DM chaperone in humans. Upon achieving a stable peptide-MHC-II confirmation, the complex is trafficked to the plasma membrane for presentation to antigen-specific T-cell receptors (TCRs), initiating CD4+ T-cell clonal expansion, macrophage activation, and downstream B-cell antibody secretion (see fig. 1) (1–3).</p>

<p>The structural biophysics governing peptide binding within the MHC-II groove dictates individual immunogenic specificities. Unlike MHC Class I molecules, the antigen-binding cleft of MHC-II is open at both termini, allowing the accommodation of longer peptides of variable dimensions (typically 13 to 25 amino acids). Stability is primarily mediated by a highly conserved 9-amino-acid core region that docks directly into corresponding polymorphic pockets within the groove, conventionally designated as the P1, P4, P6, and P9 anchor residues. Because of the open architecture of the cleft, the physical N-terminus of an eluted peptide rarely corresponds to the P1 anchor pocket. Furthermore, peptide flanking regions (PFRs) extending outside the core binding groove contribute auxiliary hydrogen bonding and steric interactions with the MHC-II backbone, substantially modulating binding affinity. Given that human MHC-II genes (HLA-DR, -DQ, and -DP) are among the most highly polymorphic loci in the human genome, distinct allelic variants exhibit highly individualized peptide-binding repertoires, complicating the precise mapping of dominant epitopes across populations (3,4).</p>

<p>To overcome the experimental bottlenecks associated with mapping these highly polymorphic interactions, machine learning frameworks have emerged as vital computational tools for predicting immunogenic MHC-II epitopes, facilitating high-throughput neoantigen selection and multi-pathogen vaccine design. A prominent state-of-the-art tool is NetMHCIIpan, which models presentation capacity by integrating data on chemical binding affinities with mass spectrometry-derived Eluted Ligand (EL) datasets (5). Eluted ligand workflows involve the inoculation of professional antigen-presenting cells (APCs) with target antigens, subsequent cell lysis, and the immunopurification of intact peptide-MHC-II complexes using allele-specific monoclonal antibodies. The bound ligands are subsequently isolated via mild acid elution and sequenced using liquid chromatography-tandem mass spectrometry (LC-MS/MS). These empirical datasets train the underlying neural network framework, NNalignMA, enabling the algorithm to deconvolve multi-allelic mass spectrometry signals and assign individual mass spectra to specific MHC-II restrictions. The predictive output yields an Eluted Ligand (EL) score, which reflects the localized probability of presentation, alongside a relative Percentile Rank (%Rank) that normalizes these scores against a comprehensive background of random natural peptides to mitigate intrinsic allelic bias (5,6). Leveraging these predictive platforms, recent experimental work by Ciacchi et al. identified functional cross-reactive T-cell responses directed against a highly conserved cholesterol-dependent cytolysin epitope shared among distinct bacterial families, raising the fundamental evolutionary question of whether hosts exert uniform selective pressures on highly cross-recognized bacterial MHC-II epitopes (7).</p>

<p>In this investigation, we systematically mapped the landscape of shared MHC-II epitopes across a panel of prominent human respiratory pathogens—<i>Bordetella pertussis</i>, <i>Haemophilus influenzae</i>, <i>Streptococcus pneumoniae</i>, and <i>Streptococcus pyogenes</i>—to identify highly conserved candidate peptides suitable for ex vivo validation. Our workflow involved a comprehensive in silico peptidome dissection, slicing overlapping peptide n-mers (n = 12 to 21 in steps of 3) from the proteomes of each pathogen to identify shared sequence configurations. The presentation likelihood of these overlapping candidates was quantified across a panel of human HLA-DR alleles using NetMHCIIpan to evaluate candidate viability in peripheral blood mononuclear cells (PBMCs).</p>

<p>From an evolutionary perspective, shared epitopes present a distinct immunological target; if a single sequence is maintained across multiple pathogenic species, host immune surveillance is theoretically more likely to encounter and recognize it. This model predicts that host immune systems apply a counter-selective pressure that drives mutations within these shared regions to evade presentation. Consequently, an evolutionary trade-off should manifest as a negative correlation between the cross-species conservation (occurrence frequency) of a peptide and its predicted presentation capacity (EL score). We tested this hypothesis by correlating the proteomic occurrence metrics of individual $n$-mers with their respective NetMHCIIpan outputs. Additionally, we cross-referenced our datasets against a baseline of empirically validated, historical T-cell epitopes from <i>B. pertussis</i> to determine whether known immunogenic sequences exhibit distinct conservation profiles relative to the unselected background peptidome (8).</p>

<p>Concurrently, this hypothesized immune-driven selection operates against a severe biophysical constraint: highly expressed housekeeping and structural proteins face stringent structural requirements to maintain cellular viability, making them highly conserved across deep evolutionary lineages. This purifying metabolic selection could effectively neutralize immune-driven evasion mechanisms, preserving cross-reactive epitopes within essential genes. Here, we isolated 136 identical overlapping peptides residing within core housekeeping genes across the four target pathogens. To evaluate these competing evolutionary forces, our figures and tables categorize these interactions through standardized quantitative descriptors, defined as follows:
<ul>
  <li><b>EL Score (Eluted Ligand Score):</b> The continuous probability value tracking the likelihood of antigen processing and surface presentation.</li>
  <li><b>Percentile (%Rank):</b> The normalized binding rank of a target sequence evaluated against a large pool of random natural human peptides, where a standard threshold of %Rank &le; 2.0% denotes functional restriction.</li>
  <li><b>Core Peptide Alignment:</b> The predicted 9-mer core motif localized directly within the P1–P9 anchor pockets.</li>
  <li><b>Proteome Combination:</b> The specific subset of bacterial pathogens sharing the identical $n$-mer sequence.</li>
</ul>
Our analytical comparisons indicate that correlations between cross-species peptide occurrence and predicted eluted ligand scores remain exceptionally weak and exhibit considerable fluctuation across different HLA-DR alleles. Finally, the relative distribution profiles separating validated <i>B. pertussis</i> immunogens from the broader, unselected bacterial peptidome were quantified to evaluate potential signs of directional selection.</p>

<p>To contextualize these shared epitope profiles within the evolutionary framework of the selected pathogens, a Neighbor-Joining phylogenetic tree was constructed in <b>Figure 4</b>. This evolutionary model was derived by extracting highly conserved housekeeping marker loci (including <i>rpoB</i> and <i>gyrB</i>) across all four bacterial proteomes, executing a multiple sequence alignment (MSA) via Clustal Omega to compute pairwise sequence identity distances, and rendering the definitive divergence branches through iTOL (9). This phylogenetic baseline provides the necessary framework to determine whether the shared epitope pools identified in <b>Table 1</b> represent a consequence of recent vertical descent or vertical/horizontal evolutionary inheritance, or if they reflect isolated functional convergence within conserved protein domains.</p>

# 3. Methods

3.1 Computational Workflow and Data ProcessingThe analysis was structured into a four-stage computational pipeline (Figure 2) designed to identify cross-reactive, immunogenic T-cell epitopes across four respiratory bacterial pathogens: Bordetella pertussis (B1917), Haemophilus influenzae (R2866), Streptococcus pneumoniae (BHN418), and Streptococcus pyogenes (emm75).

<b>Stage 1:</b> Proteome Acquisition and N-mer ExtractionProteome datasets were acquired from UniRef (12). To capture the full repertoire of potential T-cell targets, each protein sequence was systematically processed using a sliding window algorithm. This generated contiguous peptide n-mers of varying lengths (N=12, 15, 18, and 21) by shifting the window in single-residue increments across each protein sequence. This approach ensures the inclusion of overlapping fragments, preserving the structural diversity necessary for comprehensive epitope mapping.

<b>Stage 2:</b> Structural Data CompilationExtracted n-mers were compiled into a structured data frame. To ensure the integrity of the downstream analysis, we filtered the dataset to remove redundant matches resulting from homologous regions within the same protein. We constructed a unique identifier dictionary for each peptide sequence, mapping it to its respective source proteins and species. This normalization step ensured that each distinct n-mer was treated as a single entity, providing a clean, non-redundant dataset for frequency analysis.

<b>Stage 3:</b> Cross-Species Overlap and Distribution MappingIn this stage, we assessed the frequency and distribution of each unique n-mer across the four target proteomes. By comparing peptide occurrence across distinct species, we identified highly conserved sequences. We excluded self-comparisons—where a peptide matches only within its own proteome—to isolate only those peptides present in two or more bacterial species. This provided a comprehensive map of shared peptide sequences, serving as the basis for cross-reactivity assessment.

<b>Stage 4:</b> Immunogenicity Prediction and Statistical ValidationThe final stage integrated these conserved candidates with MHC-II binding affinity predictions. Filtered n-mers were processed via NetMHCIIpan-4.0 (5) in contextless mode to evaluate intrinsic binding propensity. Candidates were selected using an eluted ligand (EL) %Rank threshold of $\le 2.0\%$, normalizing affinity predictions across diverse allelic variants. Finally, we performed statistical analyses to correlate the degree of species-specific conservation with predicted EL scores. Known T-cell epitopes were cross-referenced against these filtered candidates to quantify the fraction of shared peptides and determine their potential biological significance as cross-protective vaccine targets.

The shared peptides between species are highly correlated with their evolutionary distance. The highest frequency of shared peptides was observed between *S. pneumoniae* and *S. pyogenes* (31,181 unique 15-mers), which is consistent with their status as the closest relatives in this study. The second most frequent overlap was observed between *B. pertussis* and *H. influenzae* (2,094 shared 15-mers); while these species cluster together in our analysis, their phylogenetic distance is significantly greater than that of the Streptococci. All other pairwise species combinations shared relatively few peptides, ranging from 310 to 521 15-mers (Figure 3).

Peptides present in all four species originated from highly conserved homologous proteins, typically classified as housekeeping genes. This conservation is expected, as essential genes generally exhibit lower mutation rates compared to non-essential genes (10,11). The primary functional categories identified include RNA and protein synthesis, protein degradation and translocation, as well as proteins involved in genome repair, arrangement, and nucleotide triphosphate synthesis.

#### Comparing known T cell epitopes to the dataset

<div style="text-align: justify; margin: 5px opx; max-width: 1800px;">
To explore potential links between peptide occurrence across species and immunogenicity, we analyzed a set of known Bordetella pertussis T-cell epitopes reported by da Silva et al. (8). We assessed how many of these epitopes were also present in the other three bacterial species and compared this to the broader B. pertussis peptidome. Four of the known epitopes were absent from our dataset, likely due to strain differences: the known epitopes originate from the Tahoma I strain, whereas our peptidome data derive from the B1917 strain.

Among the remaining epitopes, 0.48% (2 out of 414) were shared with one or more of the other species. For comparison, 0.23% (2,283 out of 999,995) of all B. pertussis peptides were shared. Although the fraction of shared epitopes is slightly higher than that of the general peptidome, the small number of known epitopes limits the statistical power of this comparison.

The two shared known epitopes belong to proteins with essential cellular functions: chaperonin GroEL (WP_010931507.1), which assists in protein folding, and elongation factor G (NP_882120.1), which is involved in RNA synthesis. These findings suggest that peptides conserved across species may originate from highly conserved, functionally critical proteins, consistent with the patterns observed in the broader peptidome analysis.

# 5. Discussion

<div style="text-align: justify; margin: 5px opx; max-width: 1800px;">
The objective of this project was to investigate the potential selection of cross-reactive MHC-II epitopes, which involved the generation of peptides from four bacterial proteomes: Bordetella pertussis, Haemophilus influenzae, Streptococcus pneumoniae, and Streptococcus pyogenes. We sliced peptide n-mers and found 136 15 to 21-mer peptides that are shared across all four species. These shared peptides are mostly comprised of housekeeping proteins involved in RNA and protein synthesis, protein degradation, translocation and DNA repair. However, we did not find a correlation between their occurrence in the species and their predicted MHC-II presentation.
The shared peptides between species could have also been done by searching within homologous proteins to reduce the computational effort. Determining for homologous proteins may be more laborious in manual annotations. However, only after conducting this study we found out that this is a viable alternative. Furthermore, it would not be possible to correlate the numbers of overlaps per n-mer and the predicted score if only homologous proteins are considered.
<br><br>
Another aspect that could be improved is the identification of shared peptides between the species. Currently, the shared peptides generated in this report are identical n-mers. It would be ideal to allow for more flexibility, especially in the flanking region. Substituting a residue for another one with similar size and charge could form almost the same interactions and therefore generate a good binder as well, even though the sequences are not identical. Furthermore, the binding core of the MHC-II molecule is only interacting with specific residues, pointing towards the MHC-II binding groove.
<br><br>
The generation of candidate epitopes across multiple peptide lengths (15–21 residues) initially introduced redundancy, as shorter sequences were often embedded within longer ones. However, by clustering overlapping peptides and reanalyzing them using NetMHCIIpan, we identified the optimal length for each candidate, thereby refining the prediction quality. This approach leverages the implicit cleavage biases present in the eluted ligand training data, allowing for a more accurate ex vivo assessment of epitope presentation. Notably, none of the 15-mer peptides met the presentation threshold, while 13 peptides of lengths 18 and 21 exceeded the default cutoff of >2% percentile. This suggests that longer peptides, which incorporate flanking residues as cleavage context, may better capture the biological relevance of epitope processing and presentation.
<br><br>
The refined analysis underscores the importance of peptide context in epitope prediction. While the initial approach generated a broad set of candidates, the incorporation of cleavage site context and length optimization significantly enhanced the reliability of predictions. The predominance of 18- and 21-mer peptides among the top candidates suggests that these lengths better align with the biological constraints of antigen processing. Future work could further explore the impact of species-specific protease preferences on epitope generation, potentially refining the selection criteria for cross-species candidates.
<br><br>
Additionally, since the overlaps were found in housekeeping genes that tend to be strongly conserved, the epitope may also be present in humans. If the host and the pathogen share epitopes, there may not be T cells reacting to these epitopes due to negative selection of the T cells or worse, T cells reacting against human epitopes, generating autoimmunity. This is further supported by the association of certain MHC-II allelic variants with auto-immune diseases like type 1 diabetes or multiple sclerosis (11). The issue can be overcome by comparing the obtained candidate peptides against the human proteome and filtering the shared peptides out.
<br><br>
Finally, we examined whether there is a correlation between the number of times an n-mer is present in the species and its eluted ligand score. The assessment did not return any strong correlations. One possible explanation is that the evolutionary fitness of these proteins weighs heavier than the pressure to mutate common epitopes. Additionally, the training data of NetMHCIIpan is mostly focused on good epitopes, whereas the prediction of low scores/bad binders is not as reliable as for high scores/good binders.
<br><br>
Since our analysis did not return any strong correlations, we looked at known epitopes of B. pertussis and compared them to the fraction of peptides present in the other species. We found that 0.48 % of the known epitopes of B. pertussis are shared between the species while 0.23 % of the peptides are shared. This was a surprising result, as one would expect that if there was a selective pressure to mutate the epitopes, the number of shared peptides would be larger. However, the analysis is limited by the small sample size of the known epitopes (only 414 epitopes were assessed). A better approach would be to use the known epitopes of B. pertussis from the Immune Epitope Database (IEDB) as this would generate a bigger repertoire to investigate.
<br><br>
In conclusion, we developed a method to identify shared epitopes for MHC-II presentation across different bacterial species. We identified shared candidates between Bordetella pertussis, Haemophilus influenzae, Streptococcus pneumoniae, and Streptococcus pyogenes that can now be examined ex vivo.

### Pipeline

````python
# ============================================================
# Directory Routing Configuration
# ============================================================
DATA_DIR = Path(r"/home/user/netMHCIIpan-4.0/proteomes") 
output_pep_dir = Path(r"/home/user/netMHCIIpan-4.0/output/peptide_files")
output_pep_dir.mkdir(parents=True, exist_ok=True)

# Create a dedicated directory for the Look-Up Tables (LUTs)
LUTS_DIR = Path(r"/home/user/netMHCIIpan-4.0/output/LUTs")
LUTS_DIR.mkdir(parents=True, exist_ok=True)

peptide_lengths = range(12, 22, 3) 

def process_fasta_to_df(fasta_path, species_name, n):
    all_nmers = []
    with gzip.open(fasta_path, "rt") as handle:
        for record in SeqIO.parse(handle, "fasta"):
            seq_str = str(record.seq)
            # Efficient sliding window across the protein sequence
            all_nmers.extend([seq_str[i:i+n] for i in range(len(seq_str) - n + 1)])
    
    counts = pd.Series(all_nmers).value_counts().reset_index()
    counts.columns = ['Nmer', 'Count']
    counts['Species'] = species_name
    return counts

# ============================================================
# Main Processing & Generation Loop
# ============================================================
for n in peptide_lengths:
    print(f"\n--- Processing length: {n} ---")
    all_dfs = []
    
    for fasta_path in DATA_DIR.glob("*.gz"):
        # Note: If filename is 'pathogen.fasta.gz', .stem evaluates to 'pathogen.fasta'
        # To get a completely clean species name without extensions, you can use:
        species_name = fasta_path.name.split('.')[0]
        
        try:
            all_dfs.append(process_fasta_to_df(fasta_path, species_name, n=n))
        except Exception as e:
            print(f"Error {fasta_path.name}: {e}")

    if all_dfs:
        all_data = pd.concat(all_dfs, ignore_index=True)
        
        # --------------------------------------------------------
        # NEW: SYSTEMATIC LOOK-UP TABLE (LUT) GENERATION
        # --------------------------------------------------------
        print(f"   -> Generating structural LUT files for N = {n}...")
        unique_species = all_data['Species'].unique()
        
        # 1. Generate proteome_lut_{n}.csv 
        proteome_lut = pd.DataFrame({
            'UPI': unique_species,
            'Proteome': unique_species
        })
        proteome_lut.to_csv(LUTS_DIR / f"proteome_lut_{n}.csv", index=False)
        
        # 2. Generate protein_lut_{n}.csv (Acts as structural constraint mapping)
        protein_lut = pd.DataFrame({
            'UPI': unique_species,
            'Protein': unique_species
        })
        protein_lut.to_csv(LUTS_DIR / f"protein_lut_{n}.csv", index=False)
        
        # 3. Generate nmer_table_{n}.csv 
        # Formatted line-by-line as a literal Python string tuple: ('NMER', 'SPECIES')
        # Optimized string vectorization to prevent loop slowdowns
        nmer_tuples = ("('" + all_data['Nmer'] + "', '" + all_data['Species'] + "')\n").tolist()
        with open(LUTS_DIR / f"nmer_table_{n}.csv", "w") as f_lut:
            f_lut.writelines(nmer_tuples)
            
        print(f"   -> LUTs successfully exported to: {LUTS_DIR}")
        # --------------------------------------------------------

        # Original NetMHCIIpan peptide formatting export logic
        for species, group in all_data.groupby("Species"):
            n_part = n
            pep_filename = output_pep_dir / f"{species}N{n_part}.pep"
            fasta_filename = output_pep_dir / f"{species}N{n_part}.fasta"
            
            group[['Nmer', 'Count']].to_csv(pep_filename, sep=' ', index=False, header=False)
            with open(fasta_filename, "w") as f:
                for _, row in group.iterrows():
                    f.write(f">{int(row['Count'])}_{species}\n{row['Nmer']}\n")
            print(f"   -> Exported {species} (N{n_part}) input sets")
````

#### netMHCIIpan

This script was used to automate and run several alleles in parallel. It uses the reference
[HLA types](https://drive.google.com/file/d/1oN4yBsTfykCLHsq_NGeELTwOylM9B2XM/view?usp=sharing) (13). The script was run with netMHCIIpan4.0 on

```bash
#!/bin/bash

# 1. Setup Directories and Files
allele_file="hla_ref_set.class_ii.txt"
fasta_dir="output/pepfiles"
# Use a consistent path. Ensure this folder exists or is writable.
output_dir="output/netmhciipan_results"

# 2. Prerequisites
if [[ ! -f "$allele_file" ]]; then
  echo "Error: Allele file $allele_file not found!"
  exit 1
fi

if [[ ! -d "$fasta_dir" ]]; then
  echo "Error: Directory $fasta_dir not found!"
  exit 1
fi

mkdir -p "$output_dir"

# 3. Resource Allocation
total_cores=$(nproc)
# Using 50% of cores to avoid memory bottlenecking with netMHCIIpan
max_parallel_jobs=$((total_cores / 2))
[[ $max_parallel_jobs -lt 1 ]] && max_parallel_jobs=1

# 4. Processing Function
process_file() {
  local fasta_file="$1"
  local allele="$2"
  
  export NETMHCIIpan="/home/user/netMHCIIpan-4.0"
  export TMPDIR="/tmp"

  local base_name=$(basename "$fasta_file" | cut -f 1 -d '.')
  local output_xls="$output_dir/${base_name}_${allele}.xls"
  local output_txt="$output_dir/${base_name}_${allele}.txt"

  # Force the "SUBMITTING" message to print to stderr so it bypasses buffering
  echo "[SUBMITTING] Job: $base_name | Allele: $allele" >&2

  # Redirect perl output to the text file
  perl "$NETMHCIIpan/NetMHCIIpan-4.0.pl" \
    -a "$allele" \
    -f "$fasta_file" \
    -inptype 0 \
    -context 1 \
    -filter 1 \
    -rankF 2 \
    -s 1 \
    -xls 1 \
    -xlsfile "$output_xls" > "$output_txt" 2>&1

  # Final status update
  if [[ -f "$output_xls" && -s "$output_xls" ]]; then
      echo "[SUCCESS] Finished: $base_name | Allele: $allele" >&2
  else
      echo "[ERROR] Failed: $base_name | Allele: $allele (Check $output_txt)" >&2
  fi
}

export -f process_file
export output_dir NETMHCIIpan

# 5. Execute in Parallel
# We use ::: (files) :::: (alleles) to create a Cartesian product (every file x every allele)
fasta_files=("$fasta_dir"/*.fasta)

parallel --jobs "$max_parallel_jobs" --eta --progress \
  process_file {1} {2} ::: "${fasta_files[@]}" :::: "$allele_file"

#### Occurences across Species

````python
# ============================================================
# Configuration
# ============================================================
# Windows Native Path (Uncomment if running in Windows Jupyter)
LUTS_DIR = Path(r"\\wsl.localhost\Ubuntu\home\user\netMHCIIpan-4.0\output\LUTs")

# WSL Linux Path (Uncomment if running inside WSL Ubuntu)
# LUTS_DIR = Path("")


# ============================================================
# Helper Function Implementations
# ============================================================


def load_luts_csv(directory, n):
    """Loads the Look-Up Tables for a given N-mer length from the directory.

    Expects filenames like:
    - proteome_lut_{N}.csv
    - protein_lut_{N}.csv
    - nmer_table_{N}.csv
    """
    proteome_file = directory / f"proteome_lut_{n}.csv"
    protein_file = directory / f"protein_lut_{n}.csv"
    nmer_file = directory / f"nmer_table_{n}.csv"

    # Verify files exist before attempting to read
    for f in [proteome_file, protein_file, nmer_file]:
        if not f.exists():
            raise FileNotFoundError(
                f"Missing expected LUT file: {f.name} in {directory}"
            )

    print(f"Reading CSV files for N={n}...")
    proteomeLUT = pd.read_csv(proteome_file)
    proteinLUT = pd.read_csv(protein_file)

    # Read Nmer table line-by-line (stripping whitespaces and empty lines)
    with open(nmer_file, "r") as f:
        Nmer_table = [line.strip() for line in f if line.strip()]

    return proteomeLUT, proteinLUT, Nmer_table


def proteome_fetcher(unique_nmers, proteomeLUT):
    """Maps unique N-mers to their source proteomes using the proteome LUT.

    Assumes proteomeLUT contains columns: 'UPI' and 'Proteome'.
    """
    # Create a highly efficient lookup map: UPI -> Proteome Name
    upi_to_proteome = dict(zip(proteomeLUT["UPI"], proteomeLUT["Proteome"]))

    occurrences = {}
    for nmer, upi_list in unique_nmers.items():
        matched_proteomes = set()
        for upi in upi_list:
            if upi in upi_to_proteome:
                matched_proteomes.add(upi_to_proteome[upi])

        if matched_proteomes:
            # Sorting ensures ('A', 'B') and ('B', 'A') merge into the identical key combination
            occurrences[nmer] = tuple(sorted(matched_proteomes))

    return occurrences


def count_occurrences(occurrences):
    """Counts frequencies of each unique proteome tuple and formats keys cleanly for

    plotting.
    """
    raw_counts = Counter(occurrences.values())

    # Format the combinations as strings (e.g., "Human & Mouse") for X-axis labeling
    formatted_counts = {}
    for combo, count in raw_counts.items():
        combo_string = ", ".join(combo)
        formatted_counts[combo_string] = count

    return formatted_counts


# ============================================================
# Main Analysis Execution
# ============================================================

# 1. Directory Validation
if not LUTS_DIR.exists():
    print(
        "If you are executing this inside WSL, remember to swap to the /mnt/g/... path."
    )
    raise FileNotFoundError(f"Error: The directory {LUTS_DIR} does not exist.")

figures = {}
occurrence_tables = {}

# 2. Main Iteration Loop
for N in range(12, 22, 3):
    print(f"\nProcessing N = {N}")
    print("-" * 40)

    # --- FUNCTION CALL 1: Load Look-Up Tables ---
    try:
        proteomeLUT, proteinLUT, Nmer_table = load_luts_csv(LUTS_DIR, N)
    except FileNotFoundError as e:
        print(f"Skipping N={N} due to missing files: {e}")
        continue

    # --- Convert LUT table into dataframe ---
    nmer_records = []
    for item in Nmer_table:
        # Safely evaluate string representation of tuples/lists using ast
        nmer, upi = ast.literal_eval(item)
        nmer_records.append({"Nmer": nmer, "UPI": upi})

    nmer_df = pd.DataFrame(nmer_records)
    print(f"Loaded {len(nmer_df):,} N-mers")

    # --- Group N-mers by peptide ---
    unique_nmers = nmer_df.groupby("Nmer")["UPI"].apply(list).to_dict()

    # --- FUNCTION CALL 2: Map unique N-mers to their source proteomes ---
    occurrences = proteome_fetcher(unique_nmers, proteomeLUT)

    # --- FUNCTION CALL 3: Group and aggregate occurrences ---
    occurrences_count = count_occurrences(occurrences)

    # --- Convert occurrence counts into dataframe ---
    occurrence_df = pd.DataFrame(
        {
            "Proteome_Combination": list(occurrences_count.keys()),
            "Count": list(occurrences_count.values()),
        }
    )

    # Apply logarithmic scaling safely
    occurrence_df["Log10_Count"] = occurrence_df["Count"].apply(
        lambda x: math.log10(x) if x > 0 else 0
    )

    # Calculate string length of labels for sorting layout consistency
    occurrence_df["Combination_Length"] = (
        occurrence_df["Proteome_Combination"].astype(str).str.len()
    )

    # --- Sort cleanly ---
    occurrence_df = (
        occurrence_df.sort_values(
            by=["Combination_Length", "Count"], ascending=[True, True]
        ).reset_index(drop=True)
    )

    # Save dataframes to our dictionary tracking results
    occurrence_tables[N] = occurrence_df

    # --- Plot Results ---
    fig, ax = plt.subplots(figsize=(16, 9))

    # Changed to the old color code: #00afdc
    bars = ax.bar(
        occurrence_df["Proteome_Combination"].astype(str),
        occurrence_df["Log10_Count"],
        width=0.4,
        color="#00afdc",
    )

    # Place raw string count labels perfectly above the bars
    for bar, count in zip(bars, occurrence_df["Count"]):
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.05,
            f"{count:,}",
            ha="center",
            va="bottom",
            fontsize=11,
        )

    # Chart Styling
    ax.set_title(
        f"N-mer overlap distribution (N={N})",
        fontsize=16,
        fontweight="bold",
        pad=15,
    )
    ax.set_xlabel("Proteome combinations", fontsize=14, labelpad=10)
    ax.set_ylabel("log10(Number of N-mers)", fontsize=14, labelpad=10)

    plt.xticks(rotation=45, ha="right", fontsize=10)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    ax.set_axisbelow(True)  # Keeps grid layout beneath data lines

    plt.tight_layout()
    figures[N] = fig
    print(f"Finished processing and graphing N = {N}")

# Display all figures inside your Jupyter output canvas or UI window
plt.show()
````

#### Assessing a hypothetical Correlation between Peptide Conservation Across Species and Predicted MHC-II Presentation

````python
# ============================================================
# 1. Directory & Path Setup
# ============================================================
# Aligned to match the path where your .xls prediction output files reside
folder_path = Path(
    r"\\wsl.localhost\Ubuntu\home\user\netMHCIIpan-4.0\output\netmhciipan_results"
)

# ============================================================
# 2. Sequential Analysis & Table Printing Loop
# ============================================================
for length in range(15, 22, 3):
    substring = f"N{length}"

    # Instantly locate all matching spreadsheet files using .glob()
    files = list(folder_path.glob(f"*{substring}*.xls"))

    # Skip length gracefully if files haven't finished generating yet
    if not files:
        print(f"\nProcessing N = {length}")
        print("-" * 40)
        print(
            f"Skipping N={length}: No target files found matching '*{substring}*.xls' in {folder_path}"
        )
        continue

    results = []

    for file_path in files:
        try:
            # Parse individual affinity prediction matrices
            hla_type, df_subset = load_prediction(str(file_path))

            # Standardize HLA identifiers (retains only DRB... or HLA... patterns)
            match = re.search(r"(DRB.*|HLA.*)", hla_type)
            clean_hla_type = match.group(1) if match else hla_type

            # Ensure data arrays are stripped of nulls and forced to proper numerical floats
            # NOTE: If your column name is '%Rank' instead of 'Rank', pd.to_numeric handles it safely
            rank_col = "%Rank" if "%Rank" in df_subset.columns else "Rank"

            df_subset["Score"] = pd.to_numeric(
                df_subset["Score"], errors="coerce"
            )
            df_subset[rank_col] = pd.to_numeric(
                df_subset[rank_col], errors="coerce"
            )

            # CRITICAL FIX: Spearman correlation requires numeric inputs.
            # If 'Target' is a text string of names, map it to numeric lengths or numeric occurrences.
            # Assuming 'Target' needs to be analyzed by its total cross-species value counts/occurrences:
            if "Target" in df_subset.columns:
                # Map sequence strings to their numerical frequencies within this matrix pool
                occurrence_counts = df_subset["Target"].map(
                    df_subset["Target"].value_counts()
)
            else:
                # Fallback to index mapping if 'Target' column isn't found
                occurrence_counts = pd.to_numeric(
                    df_subset.iloc[:, 1], errors="coerce"
                )

            # Drop lines where conversions or numeric parses resulted in NaN
            clean_df = pd.DataFrame(
                {
                    "Occurrence": occurrence_counts,
                    "EL_Score": df_subset["Score"],
                    "Rank": df_subset[rank_col],
                }
            ).dropna()

            if len(clean_df) < 2:
                # spearmanr requires at least 2 observations to compute a coefficient
                continue

            # Compute statistical Spearman Rank Correlations
            corr_score, p_score = spearmanr(
                clean_df["Occurrence"], clean_df["EL_Score"]
            )
            corr_rank, p_rank = spearmanr(
                clean_df["Occurrence"], clean_df["Rank"]
            )

            results.append(
                (clean_hla_type, corr_score, p_score, corr_rank, p_rank)
            )

        except Exception as file_error:
            # Boundary isolation prevents single file crashes from breaking the loop execution
            print(
                f"  [Warning] Skipping error-prone file {file_path.name}: {file_error}"
            )

    # ============================================================
    # 3. Print out Copy-Paste Ready Markdown Correlation Tables
    # ============================================================
    if results:
        # Sort results alphabetically by cleaned up HLA string type
        results.sort(key=lambda x: x[0])

        print(f"\n### Correlations for length N = {length}")
        print(
            "| HLA-type | Correlation El score | p-value El score | Correlation Rank | p-value Rank |"
        )
        print(
            "| :--- | :---: | :---: | :---: | :---: |"
        )
        for hla_type, corr_score, p_score, corr_rank, p_rank in results:
            # Clean up p-value scientific notation formatting for publication limits
            p_score_fmt = f"{p_score:.4e}" if p_score < 0.0001 else f"{p_score:.4f}"
            p_rank_fmt = f"{p_rank:.4e}" if p_rank < 0.0001 else f"{p_rank:.4f}"

            print(
                f"| {hla_type} | {corr_score:.4f} | {p_score_fmt} | {corr_rank:.4f} | {p_rank_fmt} |"
            )
    else:
        print(
            f"No active data vectors could be correlated for length N = {length}."
        )
````

It is important to assess the synthesis viability of the peptide, due to the propensity of certain sequences to aggregate or autohydrolysis. (S1-3)

````python
def check_hydrophobicity(peptide:str):

    #	Ala, Ile, Leu, Met, Phe, Trp, Val
    hydrophobic = ["A", "I", "L", "M", "F", "W", "V"]
    #	Asn, Cys, Gly, Gln, Pro, Ser, Thr, Tyr, His, Lys, Arg, Asp, Glu
    hydrophilic = ["N", "C", "G","Q", "P", "S", "T", "Y", "H", "K", "R", "D", "E"]

    h_bonds = ["D", "E", "H", "K", "N", "Q", "R", "S", "T", "Y"]

    number_phobic = 0
    number_philic = 0
    number_h_bond = 0

    length = len(peptide)
    for i in range(length):

        number_phobic += 1 * (peptide[i] in hydrophobic)
        number_philic += 1 * (peptide[i] in hydrophilic)
        number_h_bond += 1 * (peptide[i] in h_bonds)

    
    print(peptide, "Hydrophobic: ", number_phobic/length*100, "%  Hydrophilic:", number_philic/length*100, "%  H-bond forming:", number_h_bond/length*100, "%")

def check_cleavage_sites(peptide:str):
    print()

    if "DP" in peptide or "DG" in peptide:
        print(peptide, "Cleavage likely because of DP or DG (hydrolysis)")
        # print(peptide.find("DP"))

    if peptide[2] == "G":

        if peptide[0] in ["G", "P"] or peptide[1] in ["G", "P"]:

            print(peptide, "Cleavege likely because of G at Pos3 and G/P at Pos1/2 (Diketopiperazine or pyroglutamic acid formation)")

# e.g. with the candidate peptides

peptides = [
    "YFVSYYDYYQPE",
    "QPLGGKAQFGGQ",
    "GQRRYVESLSAY",
    "ERIRQIEAKALR",
    "AQFGGQRFGEME",
    "ADYVKNMITGAA",
    "QRRYVESLSAYA",
    "QPLGGKAQFGGQRFG",
    "GGKAQFGGQRFGEME",
    "KAQFGGQRFGEMEVW",
    "GKAQFGGQRFGEMEV",
    "ADYVKNMITGAAQMD",
    "VKNMITGAAQMDGAI"
]

for peptide in peptides:
    check_hydrophobicity(peptide)
    check_cleavage_sites(peptide)
````
