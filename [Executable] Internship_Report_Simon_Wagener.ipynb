# 1. Abstract

<div style="text-align: justify; margin: 5px opx; max-width: 1800px;">
Cross-reactivity through CD4+ T cell mediated responses can protect humans from multiple pathogens. In this project, the extent of potential cross-reactivity in four bacterial respiratory pathogens was examined. This was done by slicing peptides from the proteomes, searching them for identical peptides and predicting the peptides's MHC-II binding with NetMHCIIpan.   <p style="font-size: 0.9em; color: #006991;"><b>Figure 1:</b> MHC-II epitope presentation. MHC class II molecules are synthesized in the endoplasmic reticulum, where they associate with the invariant chain (Ii). The Ii occupies the peptide-binding groove via a segment called CLIP, facilitating proper folding and efficient transport of MHC II from the ER to a late endosomal compartment known as the MIIC. Within this compartment, Ii is degraded by endosomal proteases, alongside internalized antigens acquired through endocytosis or phagocytosis. The CLIP segment is protected from degradation and is subsequently replaced by an antigenic peptide through the action of a specialized chaperone, HLA-DM in humans. The peptide:MHCII complex gets exported to the cell surface upon binding and presents the peptide to the T-cell receptors. (Adapted from 1)</p>
</div>

<p>The immune system is essential for recognizing and responding to pathogens, with T cells playing a particularly vital part in identifying foreign antigens presented by MHC-II molecules. MHC-II presentation starts with the uptake of extracellular pathogens into an endolysosome. There, the pathogenic proteins are cleaved into peptides. Parallel to the endocytosis, the MHC-II molecules are translated in the endoplasmatic reticulum. Here they bind the invariant chain, a pseudo peptide containing the endosomal targeting signal. Proteases cleave the invariant chain, leaving only the MHC-II binding fragment. Subsequently, the endolysosomes and MHC-II compartments fuse, which enables the substitution of the binding fragment with an antigenic peptide with the help of a dedicated chaperone called HLA-DM in humans. Upon loading, the peptide-MHC-II complex gets transported to the cell surface, where it can interact with the T cell receptor in conjunction with a CD4 receptor to activate T helper cells. The T-helper cells mediate further immune responses, e.g. phagocyte stimulation or antibody secretion via B-cell activation (see fig. 1) (1–3).</p>

<p>Crucial in the MHC-II presentation is the binding of the peptide to the binding groove of the MHC-II molecule. The binding groove structure consists of two α-helices and one β-sheet, interacting strongly with the peptide. Generally, the binding groove interacts with regularly spaced residues on the peptide, P1, P4, P6, and P9. The lengths of the binding peptides can differ, because the binding groove is open on both sides. Therefore, the first binding residue is denoted as P1, which may not be the first residue in the peptide. Additionally, the open binding groove allows for interactions of the flanking regions outside of the binding groove with the backbone. MHC-II molecules are among the most polymorphic proteins in the human proteome, leading to different presenting specificities per peptide and allelic variant (3,4).</p>

<p>Due to these complexities, machine learning algorithms gained popularity for the prediction of MHC-II presentation. Predicting immunogenicity allows for rapid screening of candidates in neoantigen studies for e.g. vaccine development. Such a tool is NetMHCIIpan, which predicts the presentation based on binding affinity of peptides to MHC-II molecules and eluted ligand data (5). The eluted ligand data is generated by inoculating professional antigen-presenting cells with peptides. Subsequently, the antigen-presenting cells are lysed, and peptide-bound MHC-II molecules are immunopurified with antibodies against the MHC-II molecules. The peptides get eluted and analyzed with liquid chromatography-mass spectrometry. The generated data is fed into the machine learning framework NNalignMA, which allows for assigning each peptide to one or several MHC-II molecules. During training, the machine learning algorithm is primed with single allelic data, such that it can distinguish between several variants of the MHC-II molecule. In later steps, it gets trained on multi-allelic data, which allows for the deconvolution of peptides that may bind to several MHC-II variants. Once trained, peptides can be analyzed with the machine learning algorithm. The result is an eluted ligand score, which is the likelihood of a peptide to be presented by MHC-II molecules (5,6). Recent research by Ciacchi et al. found cross-recognized T-cell responses against a cholesterol-dependent cytolysin epitope. This cross-recognition can in principle protect humans against several bacterial pathogens and leads to the question of whether there is evidence for selection of cross-recognized MHC-II epitopes from a set of bacteria (7).</p>

<p>In this project, we aimed to investigate the extent of shared MHC-II epitopes across multiple bacterial pathogens and generate candidate peptides to investigate ex vivo. Our broad strategy for this was to slice peptide n-mers (for n = 9 to 21 in steps of 3) from the respiratory pathogens <i>Bordetella pertussis</i>, <i>Haemophilus influenzae</i>, <i>Streptococcus pneumoniae</i>, and <i>Streptococcus pyogenes</i>. Then the identical peptides were determined and their likelihood to be presented by MHC-II molecules was predicted with NetMHCIIpan to generate candidate peptides to be tested in peripheral blood mononuclear cells (PBMC). Furthermore, there may be a selective pressure exerted by the host’s immune system against these shared epitopes. If epitopes are shared across several species, the host’s immune system is more likely to recognize them. This would effectively apply pressure to mutate these epitopes, leading to a negative correlation between the number of shared epitopes between the species and their likelihood of being to be presented by an MHC-II molecule (represented by the eluted ligand score). We investigated this by correlating the occurrences of a peptide in the four proteomes with their eluted ligand score. Additionally, we compared the fractions of known T-cell epitopes of <i>B. pertussis</i> that are shared with one of the other three species, against all the peptides from <i>B. pertussis</i>’ proteome that are shared with one of the investigated species (8).</p>

<p>Competing with this hypothesized correlation is the fact that highly expressed proteins tend to be highly conserved because efficiency is of utmost importance for these proteins. Therefore, the purifying selective pressure exhibited in these housekeeping proteins may prevent the adaption for evading cross-recognition. We found 136 overlapping peptides between these species in housekeeping genes. There are very weak correlations between the occurrence of a peptide in these species and the predicted eluted ligand score, although it fluctuates depending on the investigated allele. Additionally, the difference between the known epitopes and the whole peptidome of <i>B. pertussis</i> was assessed.</p>

# 3. Methods

The proteomes of the four bacterial species were acquired from UniRef (<i>Bordetella pertussis</i> B1917 UP000031744, <i>Haemophilus influenzae</i> R2866 UP000002314, <i>Streptococcus pneumoniae</i> BHN418 UP000017758, <i>Streptococcus pyogenes</i> emm75 UP000298711) (12). Each protein sequence within these proteomes was systematically processed to generate peptide n-mers of varying lengths. This was achieved by sliding a fixed-length window of N residues along the entire protein sequence in single-residue increments, ensuring that every possible contiguous peptide segment was captured. By considering overlapping peptides, the analysis preserves the full diversity of potential epitopes and allows for the detection of highly conserved sequences across species.

The occurrence of identical peptide n-mers was assessed across the respective proteomes. To avoid confounding by self-comparisons, searches were restricted such that each peptide was only compared against proteins from different proteomes, and matches within the same protein were excluded. Redundant matches arising from repeated sequences or homologous regions within the proteomes were filtered out by constructing a dictionary of unique n-mers, each associated with a set of protein identifiers. This ensured that each peptide sequence was represented only once per protein, providing a clean dataset for downstream analysis.

For the immunogenicity predictions, the filtered n-mers were input into NetMHCIIpan-4.0 (5). The algorithm evaluates the binding potential of each peptide to a broad panel of MHC-II allelic variants, assigning an eluted ligand score that reflects the likelihood of presentation on the cell surface. Candidate peptides were selected based on their %Rank score, retaining only those within the top 2% compared to a background distribution of random peptides, which normalizes differences in binding affinity between MHC-II alleles. Predictions for candidate peptides were performed in contextless mode to focus on the intrinsic binding propensity of the peptide sequence itself. In parallel, correlations between peptide occurrence across species and eluted ligand scores were analyzed using 21-mer peptides, which include flanking residues to account for the influence of peptide context on MHC-II binding and presentation.

This workflow allows for a comprehensive assessment of peptide conservation, potential cross-reactivity, and immunogenic potential, providing a robust basis for identifying shared epitopes that may elicit T-cell responses across multiple bacterial pathogens.

### Custom functions

````python
import gzip
from pathlib import Path
import pandas as pd
from Bio import SeqIO

# 1. Set the correct directory
DATA_DIR = DATA_DIR = Path("/home/simon/Documents/Internship CMBI") # This is where your notebook and .gz files are located

def process_fasta_to_df(fasta_path, species_name, n=12):
    """
    Parses a gzipped FASTA file and returns a DataFrame of all N-mers.
    """
    records = []
    # Use gzip.open to handle the .gz files
    with gzip.open(fasta_path, "rt") as handle:
        for record in SeqIO.parse(handle, "fasta"):
            seq_str = str(record.seq)
            # Efficient list comprehension for n-mers
            n_mers = [seq_str[i:i+n] for i in range(len(seq_str) - n + 1)]
            
            for nmer in n_mers:
                records.append({
                    "Nmer": nmer,
                    "Protein_ID": record.id,
                    "Species": species_name
                })
    return pd.DataFrame(records)

# 2. Map your files to species names (handling the spaces in filenames)
# We use the explicit path for each file to be safe
species_map = {
    "Bordetella": DATA_DIR / "bordetella.gz",
    "Haemophilus": DATA_DIR / "heam.gz",
    "Strep_pyo": DATA_DIR / "strep pyo.gz",
    "Strep_pneu": DATA_DIR / "strep pneu.gz"
}

# 3. Combine into one Master DataFrame
print("Processing files...")
all_dfs = []
for name, path in species_map.items():
    if path.exists():
        print(f"Loading {name}...")
        all_dfs.append(process_fasta_to_df(path, name))
    else:
        print(f"Error: Could not find {path}")

if all_dfs:
    all_data = pd.concat(all_dfs, ignore_index=True)
    print(f"Success! Master DataFrame created with {len(all_data)} total N-mers.")
````

This script was used to automate and run several alleles in parallel. It uses the reference

```bash
#!/bin/bash

# Input files and directories
allele_file="hla_ref_set.class_ii.txt"
pep_dir="output/pepfiles"
output_dir="output/netmhciipan_results"

# Check prerequisites
if [[ ! -f "$allele_file" ]]; then
  echo "Error: Allele file $allele_file not found!"
  exit 1
fi

if [[ ! -d "$pep_dir" ]]; then
  echo "Error: Directory $pep_dir not found!"
  exit 1
fi

mkdir -p "$output_dir"

# Determine number of cores to use (50% of total)
total_cores=$(nproc)
max_parallel_jobs=$((total_cores * 50 / 100))
if (( max_parallel_jobs < 1 )); then
  max_parallel_jobs=1
fi

# Function to process each file
process_file() {
  local pep_file=$1
  local allele=$2
  local base_name=$(basename "$pep_file" | cut -f 1 -d '.')
  local output_file="$output_dir/${base_name}_${allele}.xls"

  if [[ $base_name =~ N[0-9]+ ]]; then
    local n_part=$(echo "$base_name" | grep -o 'N[0-9]\+')
    local peptide_length_part=${n_part#N}
    local peptide_length=$((peptide_length_part - 6))
    echo "Peptide length for $base_name is $peptide_length"
  else
    echo "Could not determine peptide length from filename: $base_name"
  fi

  local input_count=$(wc -l < "$pep_file")
  netMHCIIpan -a "$allele" -f "$pep_file" -inptype 1 -context 1 -filter 1 -rankF 2 -s 1 -xls 1 -xlsfile "$output_file" > /dev/null 2>&1

  local output_count=$(awk 'NF && NR>1' "$output_file" | wc -l)
  output_count=$((output_count - 2))

  local lengths=$(awk '{print length($0)}' "$pep_file" | sort -u | tr '\n' ',' | sed 's/,$//')

  echo "File: $base_name | Allele: $allele | Peptide lengths: $lengths | Input peptides: $input_count | Output peptides: $output_count"
}

export -f process_file
export output_dir

pep_files=("$pep_dir"/*.pep)
parallel --jobs "$max_parallel_jobs" --eta --progress --colsep '\t' \
  'process_file {1} {2}' ::: "${pep_files[@]}" :::: "$allele_file"

#### Assessing a hypothetical Correlation between Peptide Conservation Across Species and Predicted MHC-II Presentation

````python
folder_path = r"\\wsl.localhost\Ubuntu\home\user\netMHCIIpan\netMHCIIpan-4.0\clean_start_9825\output\netmhciipan_results"


for length in range(15, 22, 3):
    
    substring = f"N{length}"

    files = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.xls') and substring in filename:
            files.append(os.path.join(folder_path, filename))
        results = []
        substring = f"N{length}"

    for file in files:
        hla_type, df_subset = load_prediction(file)
        EL_score = df_subset['Score'].astype(float)
        occurrence = df_subset['Target']
        rank = df_subset['Rank'].astype(float)

        corr_score, p_score = spearmanr(occurrence, EL_score)
        corr_rank, p_rank = spearmanr(occurrence, rank)

        results.append((hla_type, corr_score, p_score, corr_rank, p_rank))

    # Sort alphabetically by HLA type
    results.sort(key=lambda x: x[0])

    # Print table
    print(f"Correlations for length N = {length}")
    print("|HLA-type| Correlation El score | p-value El score | Correlation Rank | p-value Rank |")
    print("|--------|----------------------|------------------|------------------|--------------|")
    for hla_type, corr_score, p_score, corr_rank, p_rank in results:
        print(f"| {hla_type} | {corr_score:.4f} | {p_score:.4g} | {corr_rank:.4f} | {p_rank:.4g} |")
````

#### Comparing known T cell epitopes to the dataset

````python
#Load LUTs
proteomeLUT, proteinLUT, Nmer_table = load_luts_csv(LUTs, 15)
xlsx_file = r"G:\My Drive\Internship\mmc3.xlsx"

# Ensure Nmer_table is a list of tuples
if isinstance(Nmer_table, set):
    Nmer_table = list(Nmer_table)
elif isinstance(Nmer_table, dict):
    Nmer_table = list(Nmer_table.items())
elif isinstance(Nmer_table, list):
    if len(Nmer_table) > 0 and isinstance(Nmer_table[0], str):
        Nmer_table = [ast.literal_eval(item) for item in Nmer_table]

unique_Nmers = matching2(Nmer_table)
Nmers_proteome = proteome_fetcher(unique_Nmers, proteomeLUT)

#Convert to DataFrame
df = pd.DataFrame([
    {"Nmer": nmer, "Species": list(species)}
    for nmer, species in Nmers_proteome.items()
])
df["Num_species"] = df["Species"].apply(len)
df["Contains_BPE"] = df["Species"].apply(lambda s: "BPE" in s)

#BPE proteome analysis
bpe_df = df[df["Contains_BPE"]]

# Count how many BPE N-mers have 2+ species
bpe_multi = bpe_df[bpe_df["Num_species"] >= 2]
fraction_bpe_multi = len(bpe_multi) / len(bpe_df) * 100 if len(bpe_df) > 0 else 0


print(f"Fraction of BPE peptides present in one of the other species: {fraction_bpe_multi:.2f}% ({len(bpe_multi)}/{len(bpe_df)})")

#Load known epitopes
epitopes_df = pd.read_excel(xlsx_file, skiprows=1)
known_epitopes = set(epitopes_df["Peptide Sequence"].dropna().str.strip())

#All peptides but exclude only-BPE
non_bpe_df = df[~((df["Contains_BPE"]) & (df["Num_species"] == 1))]

# Which known epitopes are present in this filtered set
filtered_peptides = set(non_bpe_df["Nmer"])
shared_epitopes = known_epitopes & filtered_peptides
fraction_epitopes = len(shared_epitopes) / len(known_epitopes) * 100

print(f"Fraction of known BPE epitopes present in one of the other species: {fraction_epitopes:.2f}% ({len(shared_epitopes)}/{len(known_epitopes)})")
````

<div style="text-align: justify; margin: 5px opx; max-width: 1800px;">
To explore potential links between peptide occurrence across species and immunogenicity, we analyzed a set of known Bordetella pertussis T-cell epitopes reported by da Silva et al. (8). We assessed how many of these epitopes were also present in the other three bacterial species and compared this to the broader B. pertussis peptidome. Four of the known epitopes were absent from our dataset, likely due to strain differences: the known epitopes originate from the Tahoma I strain, whereas our peptidome data derive from the B1917 strain.

Among the remaining epitopes, 0.48% (2 out of 414) were shared with one or more of the other species. For comparison, 0.23% (2,283 out of 999,995) of all B. pertussis peptides were shared. Although the fraction of shared epitopes is slightly higher than that of the general peptidome, the small number of known epitopes limits the statistical power of this comparison.

The two shared known epitopes belong to proteins with essential cellular functions: chaperonin GroEL (WP_010931507.1), which assists in protein folding, and elongation factor G (NP_882120.1), which is involved in RNA synthesis. These findings suggest that peptides conserved across species may originate from highly conserved, functionally critical proteins, consistent with the patterns observed in the broader peptidome analysis.

# 5. Discussion

<div style="text-align: justify; margin: 5px opx; max-width: 1800px;">
The objective of this project was to investigate the potential selection of cross-reactive MHC-II epitopes, which involved the generation of peptides from four bacterial proteomes: Bordetella pertussis, Haemophilus influenzae, Streptococcus pneumoniae, and Streptococcus pyogenes. We sliced peptide n-mers and found 136 15 to 21-mer peptides that are shared across all four species. These shared peptides are mostly comprised of housekeeping proteins involved in RNA and protein synthesis, protein degradation, translocation and DNA repair. However, we did not find a correlation between their occurrence in the species and their predicted MHC-II presentation.
The shared peptides between species could have also been done by searching within homologous proteins to reduce the computational effort. Determining for homologous proteins may be more laborious in manual annotations. However, only after conducting this study we found out that this is a viable alternative. Furthermore, it would not be possible to correlate the numbers of overlaps per n-mer and the predicted score if only homologous proteins are considered.
<br><br>
Another aspect that could be improved is the identification of shared peptides between the species. Currently, the shared peptides generated in this report are identical n-mers. It would be ideal to allow for more flexibility, especially in the flanking region. Substituting a residue for another one with similar size and charge could form almost the same interactions and therefore generate a good binder as well, even though the sequences are not identical. Furthermore, the binding core of the MHC-II molecule is only interacting with specific residues, pointing towards the MHC-II binding groove.
<br><br>
The generation of candidate epitopes across multiple peptide lengths (15–21 residues) initially introduced redundancy, as shorter sequences were often embedded within longer ones. However, by clustering overlapping peptides and reanalyzing them using NetMHCIIpan, we identified the optimal length for each candidate, thereby refining the prediction quality. This approach leverages the implicit cleavage biases present in the eluted ligand training data, allowing for a more accurate ex vivo assessment of epitope presentation. Notably, none of the 15-mer peptides met the presentation threshold, while 13 peptides of lengths 18 and 21 exceeded the default cutoff of >2% percentile. This suggests that longer peptides, which incorporate flanking residues as cleavage context, may better capture the biological relevance of epitope processing and presentation.
<br><br>
The refined analysis underscores the importance of peptide context in epitope prediction. While the initial approach generated a broad set of candidates, the incorporation of cleavage site context and length optimization significantly enhanced the reliability of predictions. The predominance of 18- and 21-mer peptides among the top candidates suggests that these lengths better align with the biological constraints of antigen processing. Future work could further explore the impact of species-specific protease preferences on epitope generation, potentially refining the selection criteria for cross-species candidates.
<br><br>
Additionally, since the overlaps were found in housekeeping genes that tend to be strongly conserved, the epitope may also be present in humans. If the host and the pathogen share epitopes, there may not be T cells reacting to these epitopes due to negative selection of the T cells or worse, T cells reacting against human epitopes, generating autoimmunity. This is further supported by the association of certain MHC-II allelic variants with auto-immune diseases like type 1 diabetes or multiple sclerosis (11). The issue can be overcome by comparing the obtained candidate peptides against the human proteome and filtering the shared peptides out.
<br><br>
Finally, we examined whether there is a correlation between the number of times an n-mer is present in the species and its eluted ligand score. The assessment did not return any strong correlations. One possible explanation is that the evolutionary fitness of these proteins weighs heavier than the pressure to mutate common epitopes. Additionally, the training data of NetMHCIIpan is mostly focused on good epitopes, whereas the prediction of low scores/bad binders is not as reliable as for high scores/good binders.
<br><br>
Since our analysis did not return any strong correlations, we looked at known epitopes of B. pertussis and compared them to the fraction of peptides present in the other species. We found that 0.48 % of the known epitopes of B. pertussis are shared between the species while 0.23 % of the peptides are shared. This was a surprising result, as one would expect that if there was a selective pressure to mutate the epitopes, the number of shared peptides would be larger. However, the analysis is limited by the small sample size of the known epitopes (only 414 epitopes were assessed). A better approach would be to use the known epitopes of B. pertussis from the Immune Epitope Database (IEDB) as this would generate a bigger repertoire to investigate.
<br><br>
In conclusion, we developed a method to identify shared epitopes for MHC-II presentation across different bacterial species. We identified shared candidates between Bordetella pertussis, Haemophilus influenzae, Streptococcus pneumoniae, and Streptococcus pyogenes that can now be examined ex vivo.

#### Peptide Synthesis

It is important to assess the synthesis viability of the peptide, due to the propensity of certain sequences to aggregate or autohydrolysis. (S1-3)

````python
def check_hydrophobicity(peptide:str):

    #	Ala, Ile, Leu, Met, Phe, Trp, Val
    hydrophobic = ["A", "I", "L", "M", "F", "W", "V"]
    #	Asn, Cys, Gly, Gln, Pro, Ser, Thr, Tyr, His, Lys, Arg, Asp, Glu
    hydrophilic = ["N", "C", "G","Q", "P", "S", "T", "Y", "H", "K", "R", "D", "E"]

    h_bonds = ["D", "E", "H", "K", "N", "Q", "R", "S", "T", "Y"]

    number_phobic = 0
    number_philic = 0
    number_h_bond = 0

    length = len(peptide)
    for i in range(length):

        number_phobic += 1 * (peptide[i] in hydrophobic)
        number_philic += 1 * (peptide[i] in hydrophilic)
        number_h_bond += 1 * (peptide[i] in h_bonds)

    
    print(peptide, "Hydrophobic: ", number_phobic/length*100, "%  Hydrophilic:", number_philic/length*100, "%  H-bond forming:", number_h_bond/length*100, "%")

def check_cleavage_sites(peptide:str):
    print()

    if "DP" in peptide or "DG" in peptide:
        print(peptide, "Cleavage likely because of DP or DG (hydrolysis)")
        # print(peptide.find("DP"))

    if peptide[2] == "G":

        if peptide[0] in ["G", "P"] or peptide[1] in ["G", "P"]:

            print(peptide, "Cleavege likely because of G at Pos3 and G/P at Pos1/2 (Diketopiperazine or pyroglutamic acid formation)")

# e.g. with the candidate peptides

peptides = [
    "YFVSYYDYYQPE",
    "QPLGGKAQFGGQ",
    "GQRRYVESLSAY",
    "ERIRQIEAKALR",
    "AQFGGQRFGEME",
    "ADYVKNMITGAA",
    "QRRYVESLSAYA",
    "QPLGGKAQFGGQRFG",
    "GGKAQFGGQRFGEME",
    "KAQFGGQRFGEMEVW",
    "GKAQFGGQRFGEMEV",
    "ADYVKNMITGAAQMD",
    "VKNMITGAAQMDGAI"
]

for peptide in peptides:
    check_hydrophobicity(peptide)
    check_cleavage_sites(peptide)
````
